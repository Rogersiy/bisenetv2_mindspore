```python
import numpy as np
```

```python
tensor1 = np.fromfile("./pred/cityscapes_val_0_0.bin",dtype=np.float32).reshape(1, 19, 1024, 2048)
```

```python
tensor2 = np.fromfile("./310/cityscapes_val_0_0.bin",dtype=np.float32).reshape(1, 19, 1024, 2048)
```

```python
# 确保两个张量具有相同的形状
if tensor1.shape != tensor2.shape:
    raise ValueError("张量形状不匹配，无法比较差异。")

# 计算差异
difference = tensor1 - tensor2

# 打印差异张量
print("差异张量：")
print(difference)

# 可选：找到最大差值和平均差值
max_difference = np.max(np.abs(difference))
mean_difference = np.mean(np.abs(difference))

print("最大差值：", max_difference)
print("平均差值：", mean_difference)
```
